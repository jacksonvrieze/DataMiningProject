```
#Stanford Tagger

os.chdir("/content/drive/My Drive/Tagger")

from nltk.tag.stanford import StanfordNERTagger
from nltk.tokenize import word_tokenize

tagger = StanfordNERTagger('/content/drive/My Drive/Tagger/stanford-ner-2018-10-16/classifiers/english.all.3class.distsim.crf.ser.gz',
'/content/drive/My Drive/Tagger/stanford-ner-2018-10-16/stanford-ner.jar',
encoding='utf-8')
```

```
import re

def tagCompany(text):
    text = re.sub("[^a-zA-Z]"," ", text)
    tagged = tagger.tag(text.split())
    
    company = ''
    for word in tagged:
        if word[1] in ['ORGANIZATION']:
          company = company + '$' + word[0].upper()
          
    return (company)
```

```
companyArray = []
for index,row in news1_sam.iterrows():
    companyArray.append(tagCompany(row['Text']))
news1_sam['Company'] = companyArray
```

```
#Loughran Dictionary 

os.chdir("/content/drive/My Drive/Lexicon")


logran_dict = pd.read_csv('LoughranMcDonald_MasterDictionary_2016.csv',sep=',')

pos_lougran = []
neg_lougran = []
for index,row in logran_dict.iterrows():
    if row['Positive'] == 2009:
        pos_lougran.append(row['Word'].lower())
    if row['Negative'] == 2009:
        neg_lougran.append(row['Word'].lower())
```

```
len(pos_lougran)
```

```
len(neg_lougran)
```

```
def lougran_pos_count(text):
    words = text.split()
    pos_count = 0
    for w in words:
        if w.lower() in pos_lougran:
            pos_count = pos_count+1
    return pos_count

def lougran_neg_count(text):
    words = text.split()
    neg_count = 0
    for w in words:
        if w.lower() in neg_lougran:
            neg_count = neg_count+1
    return neg_count
```

```
# Computing Lougran Sentiment Scores
news1_sam['neg_count'] = news1_sam['Text'].apply(lougran_neg_count)
news1_sam['pos_count'] = news1_sam['Text'].apply(lougran_pos_count)
```

```
news1_sam
```

```
news1_sam['loughran_lex_score'] = (news1_sam['pos_count'] - news1_sam['neg_count'])/(news1_sam['pos_count'] + news1_sam['neg_count'])
```

```
#Sentiment Composition Lexicon of Negators, Modals, and Adverbs (SCL-NMA)

os.chdir("/content/drive/My Drive/Lexicon")

words_lex = []
score_lex = []
def lex_SCL():
    fname = 'SCL-NMA.txt'
    f = open(fname)
    for l in f.readlines():
       
        word,score = l.split('\t')
        words_lex.append(word)
        score = score.replace('\n','')
        score_lex.append(float(score))
    return words_lex,score_lex
```

```
words_lex_SCL,score_lex_SCL = lex_SCL()
```

```
#Function for computing lexicon score by using SCL

from nltk.stem.porter import *
stemmer = PorterStemmer() 

def lex_score_SCL(text):
    tokens = text.split(' ')
    score = 0
    for word in tokens:
        for index in range(0,len(words_lex_SCL)):
            if(word.isupper() == False):
                if stemmer.stem(str(word.lower())) == stemmer.stem(words_lex_SCL[index]):
                        score = score+score_lex_SCL[index]
                        break

    return score
```

```
news1_sam['lex_score_SCL'] = news1_sam['Text'].apply(lex_score_SCL)
```
